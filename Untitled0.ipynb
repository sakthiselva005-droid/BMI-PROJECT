## Working with numpy Matrices (Multidimensional Data)

This notebook demonstrates operations with NumPy matrices using excerpts from the National Health and Nutrition Examination Survey (NHANES dataset).

### 1. Download and Load Data

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns

# Define the base URL for the datasets
base_url = "https://raw.githubusercontent.com/gagolews/teaching-data/master/marek/"

# Define filenames
file_male = "nhanes_adult_male_bmx_2020.csv"
file_female = "nhanes_adult_female_bmx_2020.csv"

# Construct full URLs
url_male = base_url + file_male
url_female = base_url + file_female

print(f"Attempting to download male data from: {url_male}")
print(f"Attempting to download female data from: {url_female}")
```

#### Purpose: Download the NHANES dataset excerpts for adult males and females from the specified GitHub repository.

```python
# Download the files and save them locally
def download_file(url, filename):
    response = requests.get(url)
    response.raise_for_status() # Raise an exception for HTTP errors
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f"Downloaded {filename}")

try:
    download_file(url_male, file_male)
    download_file(url_female, file_female)
except requests.exceptions.RequestException as e:
    print(f"Error downloading files: {e}")

print("Download complete.")
```

#### Result: The two CSV files, `nhanes_adult_male_bmx_2020.csv` and `nhanes_adult_female_bmx_2020.csv`, have been successfully downloaded and saved to the current working directory.

#### Purpose: Read the downloaded CSV files into NumPy matrices named `male` and `female`.

```python
# Read the CSV files directly into NumPy arrays using numpy.genfromtxt.
# Use delimiter=',' for CSV files, and dtype=float for numerical data.
# Increased skip_header to 9 to ignore the initial nine comment/metadata lines.
# Added comments='#' to handle any other potential comment lines starting with '#'.
# Added usecols=range(7) to explicitly specify the number of columns to read, ensuring 2D output.
male_data = np.genfromtxt(file_male, delimiter=',', dtype=float, skip_header=9, comments='#', usecols=range(7))
female_data = np.genfromtxt(file_female, delimiter=',', dtype=float, skip_header=9, comments='#', usecols=range(7))

# The task states that each matrix consists of seven columns.
# We have already specified usecols=range(7) in genfromtxt, so no further slicing is needed here.
male = male_data
female = female_data

# Remove the first row if it contains all NaN values, which is consistently observed after loading.
male = male[1:]
female = female[1:]

print("Male matrix shape (after removing initial NaN row):", male.shape)
print("Female matrix shape (after removing initial NaN row):", female.shape)

# Display the first 5 rows of each matrix to verify
print("\nFirst 5 rows of male matrix (after removing initial NaN row):")
print(male[:5])
print("\nFirst 5 rows of female matrix (after removing initial NaN row):")
print(female[:5])
```

#### Result: The body measurement data for adult males and females has been loaded into two NumPy matrices, `male` and `female`, respectively. Each matrix contains seven columns corresponding to weight, standing height, upper arm length, upper leg length, arm circumference, hip circumference, and waist circumference. The initial row containing `nan` values has been successfully removed, ensuring the matrices contain only valid numerical data.

### 6. Adding Body Mass Index (BMI) to Female Data

#### Purpose: Compute the Body Mass Index (BMI) for each female participant using their weight and standing height, and append this as a new column to the `female` matrix. The previous BMI calculation was based on an uncleaned `female` matrix; therefore, this step re-calculates the BMI with the now-cleaned data.

```python
# Extract weight (column 0) and standing height (column 1) from the female matrix
female_weight_raw = female[:, 0]
female_height_cm_raw = female[:, 1]

# Filter out NaN values from weight and height for BMI calculation
# Create a boolean mask where both weight and height are not NaN
valid_mask = ~np.isnan(female_weight_raw) & ~np.isnan(female_height_cm_raw)

female_weight = female_weight_raw[valid_mask]
female_height_cm = female_height_cm_raw[valid_mask]

# Convert height from cm to meters for BMI calculation (height in meters)
female_height_m = female_height_cm / 100

# Calculate BMI: weight (kg) / (height (m))^2
female_bmi_calculated = female_weight / (female_height_m ** 2)

# Create a full BMI array, filling NaNs for invalid rows
female_bmi = np.full(female_height_cm_raw.shape, np.nan) # Initialize with NaNs
female_bmi[valid_mask] = female_bmi_calculated # Fill in calculated BMIs

# Reshape BMI to a column vector to be able to append it to the female matrix
female_bmi = female_bmi.reshape(-1, 1)

# Add the BMI as the eighth column to the female matrix
# Re-stack female matrix to ensure the original number of rows is preserved
female = np.hstack((female, female_bmi))

print("Shape of female matrix after adding BMI:", female.shape)
print("First 5 rows of female matrix with BMI (last column):")
print(female[:5])
```

### 3. Visualizing Weight Distributions (Histograms)

#### Purpose: Draw two histograms on a single plot to visualize the distribution of female and male weights, ensuring identical x-axis limits for comparison.

```python
# Extract weight data (first column, index 0) from male and female matrices
female_weights_raw = female[:, 0]
male_weights_raw = male[:, 0]

# Remove NaN values for robust calculations and plotting
female_weights = female_weights_raw[~np.isnan(female_weights_raw)]
male_weights = male_weights_raw[~np.isnan(male_weights_raw)]

# Determine common x-axis limits for both histograms, using nan-aware min/max
min_weight = np.nanmin(female_weights)
max_weight = np.nanmax(female_weights)

if not np.isnan(np.nanmin(male_weights)):
    min_weight = min(min_weight, np.nanmin(male_weights))
if not np.isnan(np.nanmax(male_weights)):
    max_weight = max(max_weight, np.nanmax(male_weights))

# Create a figure with two subplots stacked vertically
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot histogram for female weights (top subplot)
axes[0].hist(female_weights, bins=30, color='pink', edgecolor='black', alpha=0.7)
axes[0].set_title('Distribution of Female Weights')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.75)

# Plot histogram for male weights (bottom subplot)
axes[1].hist(male_weights, bins=30, color='lightskyblue', edgecolor='black', alpha=0.7)
axes[1].set_title('Distribution of Male Weights')
axes[1].set_xlabel('Weight (kg)')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', alpha=0.75)

# Set identical x-axis limits for both subplots, ensuring valid limits
if not np.isnan(min_weight) and not np.isinf(min_weight) and \
   not np.isnan(max_weight) and not np.isinf(max_weight):
    axes[0].set_xlim(min_weight, max_weight)
    axes[1].set_xlim(min_weight, max_weight)
else:
    print("Warning: Could not determine valid axis limits due to NaN/Inf values. Using default limits.")

plt.tight_layout()
plt.show()
```

#### Result: The histograms illustrate the weight distributions for adult males and females. Both distributions appear somewhat bell-shaped, though the male weight distribution seems to be shifted to higher values compared to females, and potentially exhibits a wider spread.

### 4. Comparing Weight Distributions (Box Plot)

#### Purpose: Generate a box-and-whisker plot to visually compare the distributions of male and female weights side-by-side.

```python
# Create a box-and-whisker plot for female and male weights
plt.figure(figsize=(8, 6))
plt.boxplot([female_weights, male_weights], labels=['Female Weights', 'Male Weights'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            flierprops=dict(marker='o', markerfacecolor='red', markersize=5,
            linestyle='none'))

plt.title('Box Plot of Female vs. Male Weights')
plt.ylabel('Weight (kg)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Result: The box plot clearly illustrates the differences in weight distributions between females and males. The median weight for males appears higher than for females, and the interquartile range (IQR), represented by the box, is also generally higher for males, indicating a greater spread in the central 50% of male weights. Outliers are visible for both groups, represented by individual points outside the whiskers. The female distribution appears to be less spread out than the male distribution.

### 5. Numerical Aggregates of Weights

#### Purpose: Compute various numerical aggregates (measures of location, dispersion, and shape) for both female and male weight distributions to provide a quantitative summary.

```python
# Extract raw weight data (first column, index 0) from male and female matrices
female_weights_raw = female[:, 0]
male_weights_raw = male[:, 0]

# Filter out NaN values for numerical aggregates
female_weights = female_weights_raw[~np.isnan(female_weights_raw)]
male_weights = male_weights_raw[~np.isnan(male_weights_raw)]

# Calculate numerical aggregates for female weights
female_mean = np.nanmean(female_weights)
female_median = np.nanmedian(female_weights)
female_std = np.nanstd(female_weights)
female_min = np.nanmin(female_weights)
female_max = np.nanmax(female_weights)
female_q1 = np.nanpercentile(female_weights, 25)
female_q3 = np.nanpercentile(female_weights, 75)
female_iqr = female_q3 - female_q1

# Calculate numerical aggregates for male weights
male_mean = np.nanmean(male_weights)
male_median = np.nanmedian(male_weights)
male_std = np.nanstd(male_weights)
male_min = np.nanmin(male_weights)
male_max = np.nanmax(male_weights)
male_q1 = np.nanpercentile(male_weights, 25)
male_q3 = np.nanpercentile(male_weights, 75)
male_iqr = male_q3 - male_q1

print("Female Weights Aggregates:")
print(f"  Mean: {female_mean:.2f} kg")
print(f"  Median: {female_median:.2f} kg")
print(f"  Standard Deviation: {female_std:.2f} kg")
print(f"  Min: {female_min:.2f} kg")
print(f"  Max: {female_max:.2f} kg")
print(f"  1st Quartile (Q1): {female_q1:.2f} kg")
print(f"  3rd Quartile (Q3): {female_q3:.2f} kg")
print(f"  Interquartile Range (IQR): {female_iqr:.2f} kg")

print("\nMale Weights Aggregates:")
print(f"  Mean: {male_mean:.2f} kg")
print(f"  Median: {male_median:.2f} kg")
print(f"  Standard Deviation: {male_std:.2f} kg")
print(f"  Min: {male_min:.2f} kg")
print(f"  Max: {male_max:.2f} kg")
print(f"  1st Quartile (Q1): {male_q1:.2f} kg")
print(f"  3rd Quartile (Q3): {male_q3:.2f} kg")
print(f"  Interquartile Range (IQR): {male_iqr:.2f} kg")
```

#### Result: The numerical aggregates confirm the visual observations from the plots. Males generally have higher mean and median weights, indicating a shift towards heavier individuals. The standard deviation and interquartile range are also larger for males, suggesting greater variability and dispersion in their weight distribution compared to females. Both distributions appear somewhat symmetric around their medians, without strong indications of extreme skewness from these basic measures, though more advanced skewness tests could provide further insight.
